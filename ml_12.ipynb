```python
import numpy as np
import math
import matplotlib.pyplot as plt 
import sklearn.datasets
import sklearn.svm
import sklearn.manifold
import sklearn.decomposition

def read_data(): 
    x = [] 
    with open('iris.data', 'r') as f: 
        for line in f: 
            if line == '\n': 
                continue 
            a, b, c, d, e = line.split(',') 
            if e == 'Iris-setosa\n': 
                e = 1
            elif e == 'Iris-versicolor\n': 
                e = 2
            else: 
                e = 3
            x.append([float(a), float(b), float(c), float(d), float(e)]) 
    return x

data = np.array(read_data())
x = data[:,:4]
result_tsne = sklearn.manifold.TSNE(perplexity=30, learning_rate=30,n_iter=5000).fit_transform(x.copy())
result_pca = sklearn.decomposition.PCA(2).fit_transform(x.copy())

setosa, = plt.plot(result_tsne[:50,0],result_tsne[:50,1],"r.",label = "setosa")
versicolor, = plt.plot(result_tsne[50:100,0],result_tsne[50:100,1],"g.",label = "versicolor")
virginica, = plt.plot(result_tsne[100:150,0],result_tsne[100:150,1],"b.",label = "virginica")
plt.title("TSNE")
plt.legend(handles = [setosa,versicolor,virginica])

plt.figure(2)

setosa, = plt.plot(result_pca[:50,0],result_pca[:50,1],"r.",label = "setosa")
versicolor, = plt.plot(result_pca[50:100,0],result_pca[50:100,1],"g.",label = "versicolor")
virginica, = plt.plot(result_pca[100:150,0],result_pca[100:150,1],"b.",label = "virginica")
plt.title("PCA")
plt.legend(handles = [setosa,versicolor,virginica])

plt.show()

```
