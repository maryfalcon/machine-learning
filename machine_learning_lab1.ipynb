```python
import random 
import math
import matplotlib.pyplot as plt 

def generate_points(m, x_points, y_points, f_points, q_side): 
    for i in xrange(0, m): 
        x_points.append(random.uniform(0,1)) 
    for i in xrange(0, m): 
        y_points.append(random.uniform(0,1)) 
    for i in xrange(0, m): 
        f_points.append(y_points[i] <= q_side and x_points[i] <= q_side) 

def get_rectangle(m, x_points, y_points, f_points):
    xmin = 10
    xmax = -10
    ymin = 10
    ymax = -10
    for i in xrange(0, m): 
        if(f_points[i] == True):
            if(xmin > x_points[i]):
                xmin = x_points[i]
            if(xmax < x_points[i]):
                xmax = x_points[i]
            if(ymin > y_points[i]):
                ymin = y_points[i]
            if(ymax < y_points[i]):
                ymax = y_points[i]
    if(xmin == 10):
        return 0,0,0,0
    return xmin, xmax, ymin, ymax

def calc_true_risk(q_side, h_side1, h_side2):
    return (q_side * q_side-h_side1 * h_side2)

def rectangle_classifier(m):
    x_points = [] 
    y_points = [] 
    f_points = []
    q_side = math.sqrt(0.5)
    generate_points(m, x_points, y_points, f_points, q_side) 
    xmin, xmax, ymin, ymax  = get_rectangle(m, x_points, y_points, f_points);
    trueRisk = calc_true_risk(q_side, xmax-xmin, ymax-ymin)
    return trueRisk


true_risks = []
true_risks_i = []
m = 1000
for i in xrange(1, m):  
    true_risks.append(rectangle_classifier(i))
    true_risks_i.append(i)
plt.plot(true_risks_i, true_risks, 'r') 
plt.xlabel('m')
plt.ylabel('True risk')
plt.show()
for i in xrange(1, m):  
    print i, true_risks[i-1]
```
